```python
import pandas as pd

file_path = r"data_clean_fix.csv"
df = pd.read_csv(file_path, delimiter=",")

display(df)
```

One-Hot Encoding

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

data = pd.read_csv('data_clean_fix.csv')

corpus = data['clean_text_fix'].astype(str).tolist()

vectorizer = CountVectorizer(binary=True)  # binary=True menghasilkan one-hot encoding

one_hot = vectorizer.fit_transform(corpus)

def display_one_hot_encoding(one_hot, feature_names):
    df_one_hot = pd.DataFrame(one_hot.toarray(), columns=feature_names)
    df_one_hot.insert(0, 'clean_text_fix', data['clean_text_fix'].values)
    display(df_one_hot)
    return df_one_hot

feature_names = vectorizer.get_feature_names_out()

df_one_hot = display_one_hot_encoding(one_hot, feature_names)

df_one_hot.to_csv('one_hot_encoding_output_with_text.csv', index=False)
print("File CSV berhasil disimpan sebagai 'one_hot_encoding_output_with_text.csv'")
```

Bag-Of-Words (BOW)

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

corpus = df['clean_text_fix'].astype(str).tolist()

vectorizer = CountVectorizer()

bag_of_words = vectorizer.fit_transform(corpus)

def display_bag_of_words(bag_of_words, feature_names):
    df_bow = pd.DataFrame(bag_of_words.toarray(), columns=feature_names)
    df_bow.insert(0, 'clean_text_fix', data['clean_text_fix'].values)
    display(df_bow)
    return df_bow

feature_names = vectorizer.get_feature_names_out()

df_bow = display_bag_of_words(bag_of_words, feature_names)

df_bow.to_csv('bag_of_words_output_with_text.csv', index=False)
print("File CSV berhasil disimpan sebagai 'bag_of_words_output_with_text.csv'")
```

Pembobotan TD-IDF

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

data = pd.read_csv('data_clean_fix.csv')

corpus = data['clean_text_fix'].astype(str).tolist()

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(corpus)

def display_tfidf(tfidf_matrix, feature_names):
    df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)
    df_tfidf.insert(0, 'clean_text_fix', data['clean_text_fix'].values)
    display(df_tfidf)
    return df_tfidf

feature_names = vectorizer.get_feature_names_out()
df_tfidf = display_tfidf(tfidf_matrix, feature_names)

df_tfidf.to_csv('tfidf_output_with_text.csv', index=False)
print("File CSV berhasil disimpan sebagai 'tfidf_output_with_text.csv'")
```

```python
import nltk
nltk.download('punkt_tab')

import pandas as pd
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

corpus = df['clean_text_fix'].astype(str).tolist()
tokenized_corpus = [word_tokenize(sentence) for sentence in corpus]

model = Word2Vec(sentences=tokenized_corpus, vector_size=100, window=5, min_count=1, workers=4)

word_vectors = model.wv
words = list(word_vectors.index_to_key)

tsne = TSNE(n_components=2, random_state=0)
word_vecs_2d = tsne.fit_transform(word_vectors[words])

plt.figure(figsize=(12, 8))
plt.scatter(word_vecs_2d[:, 0], word_vecs_2d[:, 1], marker='o', color='skyblue', s=30)

for i, word in enumerate(words):
    plt.annotate(word, xy=(word_vecs_2d[i, 0], word_vecs_2d[i, 1]), fontsize=9)

plt.title("Visualisasi Word2Vec menggunakan TSNE")
plt.xlabel("TSNE Component 1")
plt.ylabel("TSNE Component 2")
plt.show()
```

# **Exploratory Data Analysis Hasil Feature-Engineering**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

corpus = df['clean_text_fix'].astype(str).tolist()

vectorizer = TfidfVectorizer()
tfidf_encoded = vectorizer.fit_transform(corpus)

feature_names = vectorizer.get_feature_names_out()
tfidf_scores = tfidf_encoded.toarray().sum(axis=0)

tfidf_dict = dict(zip(feature_names, tfidf_scores))

# Wordcloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(tfidf_dict)
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Important Words (based on TF-IDF)')
plt.show()

# Top 10 TF-IDF scores
top_10_words = sorted(tfidf_dict.items(), key=lambda item: item[1], reverse=True)[:10]

# Extract words and scores for plotting
words, scores = zip(*top_10_words)

# Create horizontal bar plot with score labels
plt.figure(figsize=(10, 6))
plt.barh(words, scores, color='skyblue')
plt.title('Top 10 TF-IDF Scores')
plt.xlabel('TF-IDF Score')
plt.ylabel('Words')
plt.gca().invert_yaxis()

# score tf-idf above each bar
for i, v in enumerate(scores):
    plt.text(v + 0.01, i, str(round(v, 3)), color='black', va='center')

plt.tight_layout()
plt.show()
```
