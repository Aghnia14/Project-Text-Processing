# **Pre-Processing**

Loading Scraped Data From Twitter

```python
import pandas as pd

file_path = r"Hasil_Scraping_Data_Kel4_23C.csv"
df = pd.read_csv(file_path, delimiter=",")

display(df)
```

Columns Filtering Raw Data

```python
df_filtered = df[['username', 'full_text']]

display(df_filtered)
```

Lower Casing

```python
df_filtered.loc[:, 'casefolded_fullText'] = df_filtered['full_text'].str.lower()

display(df_filtered)
```

Removing Mention

```python
import re

def remove_mentions(text):
    return re.sub(r'@\w+', '', text)

df_filtered['no_mentions_fullText'] = df_filtered['casefolded_fullText'].apply(remove_mentions)

display(df_filtered)
```

Removing Hashtag

```python
import re

def remove_hashtags(text):
    return re.sub(r'#\w+', '', text)

df_filtered['no_hashtag_fullText'] = df_filtered['no_mentions_fullText'].apply(remove_hashtags)

display(df_filtered)
```

Removing Retweet

```python
import re

def remove_retweets(text):
    return re.sub(r'RT\s*', '', text)

df_filtered['no_RT_fullText'] = df_filtered['no_hashtag_fullText'].apply(remove_retweets)

display(df_filtered)
```

Removing Links

```python
import re

def remove_links(text):
    url_pattern = r'http[s]?://\S+|www\.\S+'
    return re.sub(url_pattern, '', text)

df_filtered['no_url_fullText'] = df_filtered['no_RT_fullText'].apply(remove_links)

display(df_filtered)
```

Removing Punctuations

```python
import string

def remove_punctuations(text):
    return text.translate(str.maketrans('', '', string.punctuation))

df_filtered['no_punctuations_fullText'] = df_filtered['no_url_fullText'].apply(remove_punctuations)

display(df_filtered)
```

Removing Numbers

```python
import re

def remove_numbers(text):
    return re.sub(r'[0-9]+', '', text)

df_filtered['no_number_fullText'] = df_filtered['no_punctuations_fullText'].apply(remove_numbers)

display(df_filtered)
```

replace new line into space and remove character space from both left and right text

```python
def replace_newline_and_trim(text):
    text = text.replace('\n', ' ')
    return text.strip()

df_filtered['cleaned_fullText'] = df_filtered['no_number_fullText'].apply(replace_newline_and_trim)

display(df_filtered)
```

Removing All Character non alphabet

```python
import re

def remove_non_alphabet(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)

df_filtered['cleaned_fullText'] = df_filtered['cleaned_fullText'].apply(remove_non_alphabet)

display(df_filtered)
```

Encoding Character Removal

```python
def remove_encoding_artifacts(text):
    return text.encode('latin1').decode('utf-8', 'ignore')

df_filtered['cleaned_fullText'] = df_filtered['cleaned_fullText'].apply(remove_encoding_artifacts)

display(df_filtered)
```

Space Normalization

```python
import re

def normalize_spacing(text):
    text = text.strip()
    text = re.sub(r'\s+', ' ', text)
    return text

df_filtered['normalized_fullText'] = df_filtered['cleaned_fullText'].apply(normalize_spacing)

display(df_filtered)
```

Text Normalization ; Including Slang Normalization, Spelling Correction, Chat Words Removal



*   Phrase Normalization


```python
import pandas as pd
import re

phrases_df = pd.read_csv("frasa.csv")

replacement_dict = dict(zip(phrases_df['original'], phrases_df['replacement']))

def replace_phrases(text, replacement_dict):
    for original, replacement in replacement_dict.items():
        text = re.sub(r'\b' + re.escape(original) + r'\b', replacement, text, flags=re.IGNORECASE)
    return text

df_filtered['normalizedphrase_fullText'] = df_filtered['normalized_fullText'].apply(lambda x: replace_phrases(x, replacement_dict))

display(df_filtered)
```



*   Words Normalization



```python
normalization_dict = pd.read_csv("dictionary.csv")
normalization_dict = dict(zip(normalization_dict['original'], normalization_dict['replacement']))

def normalize_slang(text, normalization_dict):
    text = str(text)
    words = text.split()
    normalized_words = [normalization_dict.get(word, word) for word in words]
    normalized_words = [str(word) for word in normalized_words]
    return ' '.join(normalized_words)

df_filtered['normalizedclear_fullText'] = df_filtered['normalizedphrase_fullText'].apply(lambda x: normalize_slang(x, normalization_dict))

display(df_filtered)
```

Removing Stop Words

```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('indonesian'))

def remove_stopwords(text):
    if isinstance(text, str):  # Memastikan input adalah string
        words = text.split()  # Membagi teks menjadi kata-kata
        filtered_words = [word for word in words if word.lower() not in stop_words]  # Menghapus stopwords
        return ' '.join(filtered_words)  # Menggabungkan kata-kata yang telah difilter
    return text

df_filtered['stopwords_fullText'] = df_filtered['normalizedclear_fullText'].apply(remove_stopwords)

display(df_filtered)
```

```python
remove_words = [
    "wkwkw", "wkwkwkwkwk", "hahahaha", "mantap pak sikat", "kwkkwkw", "wkwk", "cuaks", "eh", "ehe", "hehehe",
    "wkwkwkwkwkwk", "wk", "awkwkwkwk", "honghong", "haha", "wkwkwkwk", "wkwkwk", "aahhhh", "xixixixi",
    "ckck", "awokwokwok", "hahah", "akwakwkawk", "ahahahaha", "clckckck", "ckckckck", "ttt", "ne", "ppp",
    "huhuha", "we", "ee", "nder", "nderr", "elaaaah", "loh", "nah", "hah", "ha", "woyyy", "woy", "cuy", "cuk",
    "mer","shettttt","wushwjjxjsksk","cenah","we","nder","min","cie","cebong","toh","ahahaha","ah","haha","ampamp",
    "cie","min","in","si","nya","nih","wal","wi","deh","ya","iya","sih", "sdp", "pinagkalandakan", "t", "ra", "ter",
    "h", "yo", "hm", "a", "ash", "fan", "sik", "mh", "wal", "ka", "o","ewek","si","tuh","ih","ish"
]

def remove_words_from_text(text, remove_words):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in remove_words]
    return ' '.join(filtered_words)

df_filtered['stopwords_fullText'] = df_filtered['stopwords_fullText'].apply(lambda x: remove_words_from_text(x, remove_words))

display(df_filtered)
```

Tokenizing

```python
import re

def tokenize_text(text):
    if isinstance(text, str):
        return re.findall(r'\b\w+\b', text.lower())
    return text

df_filtered['tokenized_fullText'] = df_filtered['stopwords_fullText'].apply(tokenize_text)

display(df_filtered)
```

Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemmed_wrapper(term):
    return stemmer.stem(term)

# Menginisialisasi dictionary untuk menyimpan hasil stemming
term_dict = {}
hitung = 0

for document in df_filtered['stopwords_fullText']:
    # Tokenisasi
    terms = document.split()
    for term in terms:
        if term not in term_dict:
            term_dict[term] = stemmed_wrapper(term)

print(len(term_dict))
print("------------------------")

# Melakukan stemming untuk setiap term di term_dict
for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    hitung += 1
    print(hitung, ":", term, ":", term_dict[term])

print(term_dict)
print("------------------------")

# Fungsi untuk mendapatkan term yang sudah distem
def get_stemmed_term(document):
    terms = document.split()  # Ganti ini jika diperlukan untuk tokenisasi
    return [term_dict.get(term, term) for term in terms]  # Menggunakan get untuk menghindari KeyError

df_filtered['stemmed_fullText'] = df_filtered['stopwords_fullText'].apply(lambda x: ' '.join(get_stemmed_term(x)))
display(df_filtered)
```

Lemmatization

```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')  # Untuk mendukung bahasa lebih luas jika diperlukan

wnl = WordNetLemmatizer()

def lemmatized_wrapper(term, pos='n'):
    return wnl.lemmatize(term, pos)

# Menginisialisasi dictionary untuk menyimpan hasil lemmatization
term_dict = {}
hitung = 0

for document in df_filtered['stopwords_fullText']:
    # Tokenisasi
    terms = document.split()
    for term in terms:
        if term not in term_dict:
            term_dict[term] = ' '

print(len(term_dict))
print("------------------------")

# Melakukan lemmatization untuk setiap term di term_dict (misalnya untuk nouns 'n' atau verbs 'v')
for term in term_dict:
    term_dict[term] = lemmatized_wrapper(term, 'n')
    hitung += 1
    print(f"{hitung}: {term} -> {term_dict[term]}")

print("Lemmatization selesai.")
print("------------------------")

# Fungsi untuk mendapatkan term yang sudah dilemmatized
def get_lemmatized_term(document):
    terms = document.split()  # Tokenisasi teks
    return [term_dict.get(term, term) for term in terms]  # Menghindari KeyError jika term tidak ditemukan

df_filtered['lemmatized_fullText'] = df_filtered['stopwords_fullText'].apply(lambda x: ' '.join(get_lemmatized_term(x)))

display(df_filtered)
```

# **File Csv Data Fix Clean** (setelah filter duplicate)

```python
import pandas as pd

df = pd.read_csv('data_clean.csv')

# Menghapus baris duplikat
df_cleaned = df.drop_duplicates()

# Menghapus baris yang kosong atau hanya berisi spasi pada semua kolom
df_cleaned = df_cleaned.dropna(how='any')  # Menghapus baris dengan nilai NaN
df_cleaned = df_cleaned[~df_cleaned.eq(' ').any(axis=1)]  # Menghapus baris dengan spasi kosong

df_cleaned = df_cleaned.rename(columns={'lemmatized_fullText': 'clean_text_fix'})

df_cleaned.to_csv('data_clean_fix.csv', index=False)

display(df_cleaned)
```
