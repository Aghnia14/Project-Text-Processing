```python
import pandas as pd

file_path = r"Hasil_Scrapping_Data_Kel4_23C.csv"
df = pd.read_csv(file_path, delimiter=",")

df_filtered = df[['username', 'full_text']]

display(df_filtered)
```

Lower Casing

```python
df_filtered.loc[:, 'casefolded_fullText'] = df_filtered['full_text'].str.lower()

display(df_filtered)
```

Remove Mention

```python
import re

def remove_mentions(text):
    return re.sub(r'@\w+', '', text)

df_filtered['no_mentions_fullText'] = df_filtered['casefolded_fullText'].apply(remove_mentions)

display(df_filtered)
```

Remove Hashtag

```python
import re

def remove_hashtags(text):
    return re.sub(r'#\w+', '', text)

df_filtered['no_hashtag_fullText'] = df_filtered['no_mentions_fullText'].apply(remove_hashtags)

display(df_filtered)
```

Remove Retweet

```python
import re

def remove_retweets(text):
    return re.sub(r'RT\s*', '', text)

df_filtered['no_RT_fullText'] = df_filtered['no_hashtag_fullText'].apply(remove_retweets)

display(df_filtered)
```

Remove Links

```python
import re

def remove_links(text):
    url_pattern = r'http[s]?://\S+|www\.\S+'
    return re.sub(url_pattern, '', text)

df_filtered['no_url_fullText'] = df_filtered['no_RT_fullText'].apply(remove_links)

display(df_filtered)
```

Remove Punctuations

```python
import string

def remove_punctuations(text):
    return text.translate(str.maketrans('', '', string.punctuation))

df_filtered['no_punctuations_fullText'] = df_filtered['no_url_fullText'].apply(remove_punctuations)

display(df_filtered)
```

Remove Numbers

```python
import re

def remove_numbers(text):
    return re.sub(r'[0-9]+', '', text)

df_filtered['no_number_fullText'] = df_filtered['no_punctuations_fullText'].apply(remove_numbers)

display(df_filtered)
```

replace new line into space and remove character space from both left and right text

```python
def replace_newline_and_trim(text):
    text = text.replace('\n', ' ')
    return text.strip()

df_filtered['cleaned_fullText'] = df_filtered['no_number_fullText'].apply(replace_newline_and_trim)

display(df_filtered)
```

Remove All Character non alphabet

```python
import re

def remove_non_alphabet(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)

df_filtered['cleaned_fullText'] = df_filtered['cleaned_fullText'].apply(remove_non_alphabet)

display(df_filtered)
```

```python
# Fungsi untuk membersihkan karakter yang tidak diinginkan
def remove_encoding_artifacts(text):
    return text.encode('latin1').decode('utf-8', 'ignore')

# Terapkan ke kolom yang mengandung karakter aneh
df_filtered['cleaned_fullText'] = df_filtered['cleaned_fullText'].apply(remove_encoding_artifacts)

# Tampilkan DataFrame yang sudah dibersihkan
display(df_filtered)
```

```python
# Hapus baris yang hanya berisi spasi atau kosong pada kolom 'cleaned_fullText'
df_filtered = df_filtered[~df_filtered['cleaned_fullText'].str.strip().eq("")]

display(df_filtered)
```

```python
import pandas as pd

# Memunculkan baris dengan nilai NaN atau bertipe float di kolom 'cleaned_fullText'
df_filtered_with_float_or_nan = df_filtered[df_filtered['cleaned_fullText'].isna() | df_filtered['cleaned_fullText'].apply(lambda x: isinstance(x, float))]

# Tampilkan hasilnya
display(df_filtered_with_float_or_nan)
```

```python
import pandas as pd

# Menghapus kolom yang bertipe float atau berisi nilai NaN
df_filtered = df_filtered.dropna(axis=1, how='all')  # Menghapus kolom yang seluruhnya NaN
df_filtered = df_filtered.loc[:, ~df_filtered.dtypes.isin([float])]

# Tampilkan hasil DataFrame setelah penghapusan kolom
display(df_filtered)
```

```python
import re

def normalize_spacing(text):
    # Menghapus spasi ekstra di awal dan akhir
    text = text.strip()

    # Mengganti spasi berlebih dengan satu spasi
    text = re.sub(r'\s+', ' ', text)

    return text

# Terapkan normalisasi spasi pada kolom DataFrame
df_filtered['normalized_fullText'] = df_filtered['cleaned_fullText'].apply(normalize_spacing)

# Tampilkan DataFrame yang sudah dinormalisasi
display(df_filtered)
```

Slang Normalization

```python
slang_dict = pd.read_csv("slang_dict.csv")
slang_dict = dict(zip(slang_dict['original'], slang_dict['replacement']))

def normalize_slang(text, slang_dict):
    words = text.split()
    normalized_words = [slang_dict.get(word, word) for word in words]
    return ' '.join(normalized_words)

df_filtered['normalized_fullText'] = df_filtered['cleaned_fullText'].apply(lambda x: normalize_slang(x, slang_dict))

display(df_filtered)
```

Chat Words

```python
import pandas as pd

chat_words_dict = pd.read_csv("chat_words_dict.csv")
chat_words_dict = dict(zip(chat_words_dict['singkatan'], chat_words_dict['replacement']))

def replace_chat_words(text, chat_words_dict):
    words = text.split()
    new_words = [chat_words_dict.get(word, word) for word in words]
    return ' '.join(new_words)

df_filtered['chatwords_fullText'] = df_filtered['normalized_fullText'].apply(lambda x: replace_chat_words(x, chat_words_dict))

display(df_filtered)
```

Spelling Correction

```python
spellchecker_dict = pd.read_csv("spellchecker_dict.csv")
spellchecker_dict = dict(zip(spellchecker_dict['original'], spellchecker_dict['replacement']))

def spellcheck_text(text, spellchecker_dict):
    words = text.split()
    corrected_words = [spellchecker_dict.get(word, word) for word in words]
    return ' '.join(corrected_words)

df_filtered['spellchecked_fullText'] = df_filtered['chatwords_fullText'].apply(lambda x: spellcheck_text(x, spellchecker_dict))

display(df_filtered)
```

Removal Stop Words

```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('indonesian'))

def remove_stopwords(text):
    if isinstance(text, str):  # Memastikan input adalah string
        words = text.split()  # Membagi teks menjadi kata-kata
        filtered_words = [word for word in words if word.lower() not in stop_words]  # Menghapus stopwords
        return ' '.join(filtered_words)  # Menggabungkan kata-kata yang telah difilter
    return text

df_filtered['stopwords_fullText'] = df_filtered['chatwords_fullText'].apply(remove_stopwords)

display(df_filtered)
```

```python
remove_words = [
    "wkwkw", "wkwkwkwkwk", "hahahaha", "mantap pak sikat", "kwkkwkw", "wkwk", "cuaks", "eh ehe", "hehehe",
    "wkwkwkwkwkwk", "wk", "awkwkwkwk", "honghong", "haha", "wkwkwkwk", "wkwkwk", "aahhhh", "xixixixi",
    "ckck", "awokwokwok", "hahah", "akwakwkawk", "ahahahaha", "clckckck", "ckckckck", "ttt", "ne", "ppp",
    "huhuha", "we", "ee", "nder", "nderr", "elaaaah", "loh", "nah", "hah", "ha", "woyyy", "woy", "cuy", "cuk"
]

def remove_words_from_text(text, remove_words):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in remove_words]
    return ' '.join(filtered_words)

df_filtered['stopwords_fullText'] = df_filtered['stopwords_fullText'].apply(lambda x: remove_words_from_text(x, remove_words))

display(df_filtered)
```

Tokenizing

```python
import re

def tokenize_text(text):
    if isinstance(text, str):
        return re.findall(r'\b\w+\b', text.lower())
    return text

df_filtered['tokenized_fullText'] = df_filtered['stopwords_fullText'].apply(tokenize_text)

display(df_filtered)
```

Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Membuat objek stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming
def stemmed_wrapper(term):
    return stemmer.stem(term)

# Menginisialisasi dictionary untuk menyimpan hasil stemming
term_dict = {}
hitung = 0

# Membuat dictionary term_dict dari kolom 'cleaned_fullText'
for document in df_filtered['stopwords_fullText']:
    # Tokenisasi (misalnya, memisahkan kata berdasarkan spasi)
    terms = document.split()  # Ganti ini dengan metode tokenisasi yang sesuai jika diperlukan
    for term in terms:
        if term not in term_dict:
            term_dict[term] = stemmed_wrapper(term)

print(len(term_dict))
print("------------------------")

# Melakukan stemming untuk setiap term di term_dict
for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    hitung += 1
    print(hitung, ":", term, ":", term_dict[term])

print(term_dict)
print("------------------------")

# Fungsi untuk mendapatkan term yang sudah distem
def get_stemmed_term(document):
    terms = document.split()  # Ganti ini jika diperlukan untuk tokenisasi
    return [term_dict.get(term, term) for term in terms]  # Menggunakan get untuk menghindari KeyError

# Mengupdate DataFrame dengan kolom baru 'text_stemmed'
df_filtered['stemmed_fullText'] = df_filtered['stopwords_fullText'].apply(lambda x: ' '.join(get_stemmed_term(x)))
display(df_filtered)
```

Lemmatization

```python
# Import library untuk lemmatization dengan WordNetLemmatizer
import nltk
from nltk.stem import WordNetLemmatizer

# Mengunduh WordNet
nltk.download('wordnet')
nltk.download('omw-1.4')  # Untuk mendukung bahasa lebih luas jika diperlukan

# Membuat objek WordNetLemmatizer
wnl = WordNetLemmatizer()

# Fungsi untuk melakukan lemmatization
def lemmatized_wrapper(term, pos='n'):
    return wnl.lemmatize(term, pos)

# Menginisialisasi dictionary untuk menyimpan hasil lemmatization
term_dict = {}
hitung = 0

# Membuat dictionary term_dict dari kolom 'stopwords_fullText'
for document in df_filtered['stopwords_fullText']:
    # Tokenisasi (misalnya, memisahkan kata berdasarkan spasi)
    terms = document.split()
    for term in terms:
        if term not in term_dict:
            term_dict[term] = ' '

print(len(term_dict))
print("------------------------")

# Melakukan lemmatization untuk setiap term di term_dict (misalnya untuk nouns 'n' atau verbs 'v')
for term in term_dict:
    term_dict[term] = lemmatized_wrapper(term, 'n')
    hitung += 1
    print(f"{hitung}: {term} -> {term_dict[term]}")

print("Lemmatization selesai.")
print("------------------------")

# Fungsi untuk mendapatkan term yang sudah dilemmatized
def get_lemmatized_term(document):
    terms = document.split()  # Tokenisasi teks
    return [term_dict.get(term, term) for term in terms]  # Menghindari KeyError jika term tidak ditemukan

# Mengupdate DataFrame dengan kolom baru 'lemmatized_fullText'
df_filtered['lemmatized_fullText'] = df_filtered['stopwords_fullText'].apply(lambda x: ' '.join(get_lemmatized_term(x)))

# Menampilkan DataFrame yang diperbarui
display(df_filtered)
```
